```python
#@title Dice Distribution
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import random


# Lists to store data
two_six_sided_sums = []
single_eleven_sided_rolls = []

# Simulate 10,000 throws
NUM_ROLLS = 10000
for i in range(NUM_ROLLS):
    # Rolling two 6-sided dice [1, 6]
    six_sided_dice_one = random.randint(1, 6)
    six_sided_dice_two = random.randint(1, 6)
    two_six_sided_sums.append(six_sided_dice_one + six_sided_dice_two)

    # Rolling one dice [2, 12]
    eleven_sided = random.randint(2, 12)
    single_eleven_sided_rolls.append(eleven_sided)



# Create Traces
six_sided_trace = go.Histogram(x=two_six_sided_sums, name='Two 6-sided Dice Sum')
eleven_sided_trace = go.Histogram(x=single_eleven_sided_rolls, name='Single 11-sided Dice')

# Graph figure
fig2 = make_subplots(rows=2, cols=1)
fig2.append_trace(six_sided_trace, 1, 1)
fig2.append_trace(eleven_sided_trace, 2, 1)
fig2.update_layout(bargap=0.1, title_text='Distribution after 10,000 simulated throws')
fig2.update_xaxes(dict(tickmode='linear', dtick=1))
fig2.show()
```

```python
# same but in matplotlib

import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 1, figsize=(10, 10))
axs[0].hist(two_six_sided_sums, bins=11, color='blue', edgecolor='black')
axs[0].set_title('Two 6-sided Dice Sum')
axs[0].set_xlabel('Sum')
axs[0].set_ylabel('Frequency')
axs[1].hist(single_eleven_sided_rolls, bins=11, color='red', edgecolor='black')
axs[1].set_title('Single 11-sided Dice')
axs[1].set_xlabel('Sum')
axs[1].set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```
